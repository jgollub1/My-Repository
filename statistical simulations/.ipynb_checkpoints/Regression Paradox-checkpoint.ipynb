When I helped teach stat 110 last fall, this was one of the toughest problems to explain. I won't paste the entire problem, but here's the gist of it:

Let X be the height of a randomly chosen adult man, and Y be the height of his father,
where X and Y are standardized with mean 0 and standard deviation 1. Suppose that (X,Y) 
is Bivariate Normal, with X,Y ~ N(0,1) and Corr(X,Y) = p.
...

    d) Find E(Y|X) and E(X|Y)

In the first part of the problem, we figure the slope a to be 1.3 in the equation Y = aX + B, where X represents a man's height and Y represents his father's height. I'm going to draw a simulation of this situation by simulating a linear regression with appropriate noise.

```python
# draw a bunch of men's heights (in cm) from a normal distribution
men_heights = [np.random.normal(177,7.5) for i in range(100000)]
# father's height is based off of son's height by multiplying by a and adding noise
father_heights = [np.random.normal(1.3*men_heights[i],7) for i in range(100000)]
```

```python
# checking our simulation, we observe an 81% correlation between the two groups
d = sp.pearsonr(men_heights,father_heights)
print d
```

```python
# here are side-by-side histograms of each population by height
a = pd.Series(men_heights)
a.hist()

b = pd.Series(father_heights)
b.hist()
```

```python
plt.scatter(a,b)
```

```python
sets = zip(men_heights,father_heights)
```

```python
# here, I find the expectation of the father's height, given the son is between 190 and 191cm
# and vice versa

father_exp_1 = [x for x in sets if x[0] > 190 and x[0] < 191]
men_exp_1 = [x for x in sets if x[1] > 190 and x[1] < 191]
```

```python
np.mean(men_exp_1)
```

```python
np.mean(father_exp_1)
```

```python
# we find expectations of father's heights, given the son's height for sons 
# with height 150-200cm; we are discretizing the data by mapping everything within .5 units to
# that value itself (eg, take all Y values from 149.5 < X < 150.5 to find E(Y|X=150))
father_exp_heights = {}
for i in range(150,200):
    father_exp = [x for x in sets if x[0] > (i-.5) and x[0] < (i+.5)]
    father_exp_heights[i] = np.mean(father_exp)


```

```python
# another approach to the same thing

all_heights = [np.random.multivariate_normal([177,265.5],[[56.25,45],[45,56.25]]) 
               for i in range(100000)]
```

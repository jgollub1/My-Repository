In the first part of the problem, we figure the slope a to be 1.3 in the equation Y = aX + B, where X represents a man's height and Y represents his father's height. I'm going to draw a simulation of this situation by simulating a linear regression with appropriate noise.

```python
# draw a bunch of men's heights (in cm) from a normal distribution
men_heights = [np.random.normal(177,7.5) for i in range(100000)]
# father's height is based off of son's height by multiplying by a and adding noise
father_heights = [np.random.normal(1.3*men_heights[i],7) for i in range(100000)]
```

```python
# checking our simulation, we observe an 81% correlation between the two groups
d = sp.pearsonr(men_heights,father_heights)
print d
```
